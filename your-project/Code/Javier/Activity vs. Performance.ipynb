## Analysis of activity vs. top performances

In this part, the main goal will be to observe if the athetes' best results depend on their average activity per week.

### 1- Data extraction

```python
# I first import all the necessary libraries:
import pandas as pd 
import json 
import numpy as np
```

### 2- Sports activities frequency by employment situation

```python
#I first obtain and clean the database of sports frequency by level of education and employment situation:
sports_freq_pop = pd.read_csv('../../Sources/INE/H15F3002c.csv', engine = 'python', skiprows = 6, nrows=16, index_col=0)
#I then clean the data and get only the data for the employment situation:
sports_freq_pop = sports_freq_pop.iloc[:,:5]
sports_freq_pop.rename(columns = {'Total practicaron en el �ltimo a�o': 'Yearly', 'Todos los d�as':'Daily',
       'Al menos una vez a la semana':'Weekly', 'Al menos una vez al mes': 'Monthly',
       'Al menos una vez al trimestre':'Quarterly'}, inplace = True)
sports_freq_empl = sports_freq_pop.iloc[10:]
idx = sports_freq_empl.index.str.replace('Sit. Laboral: ','')
sports_freq_empl.set_index(idx, inplace = True)
sports_freq_empl.rename(index = {'Trabajando':'Working','Parado':'Unemployed','Jubilado o retirado del trabajo':'Retired',
                                 'Estudiante':'Student','Labores del hogar':'Housework','Otros':'Others'},inplace = True)
sports_freq_empl
```

### 3- Sports activities frequency by age and sex

```python
#As before, I first obtain the database and clean the dataframe:
sports_freq_age_sex = pd.read_csv('../../Sources/INE/H15F3001c.csv', engine = 'python', skiprows = 6, index_col=0)
sports_freq_age_sex = sports_freq_age_sex.iloc[:,:5]
sports_freq_age_sex.rename(columns = {'Total practicaron en el �ltimo a�o': 'Yearly', 'Todos los d�as':'Daily',
       'Al menos una vez a la semana':'Weekly', 'Al menos una vez al mes': 'Monthly',
       'Al menos una vez al trimestre':'Quarterly'}, inplace = True)
#I then obtain a dataframe for the sex and another one for the age:
sports_freq_sex = sports_freq_age_sex.iloc[1:3]
sports_freq_sex.rename(index = {'Hombres':'Men','Mujeres':'Women'}, inplace = True)
sports_freq_age = sports_freq_age_sex.iloc[3:11]
sports_freq_age.rename(index = {'De 15 a 19 a�os':'15 - 19', 'De 20 a 24 a�os':'20 - 24', 'De 25 a 34 a�os':'25 - 34',
       'De 35 a 44 a�os':'35 - 44', 'De 45 a 54 a�os':'45 - 54', 'De 55 a 64 a�os':'55 - 64',
       'De 65 a 74 a�os':'65 - 74', 'De 75 a�os y m�s':'> 75'}, inplace = True)
sports_freq_age
```
