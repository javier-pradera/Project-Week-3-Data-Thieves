```python
import pandas as pd
import requests
import re 
import json 
from bs4 import BeautifulSoup as Soup
```

```python
url = "https://www.strava.com/pros"

#let's request the info
page = requests.get(url)
soup = Soup(page.content, "html.parser")
```

```python
#finding top countries from the pros 
top_profiles_countries = [element["alt"] for element in soup.select("div ul.pros")[0].select("div.location img")]
```

```python
countries = list(top_profiles_countries[0:500])
```

```python
#finding cities from the pros 
top_profiles_cities = [element.get_text() for element in soup.select("div ul.pros")[0].select("div.location img")]
```

```python
cities = []
for i in top_profiles_cities:
    x = re.sub('\r?\n', '', i)
    cities.append(x)

#now we hace the cities also.    
cities = cities[0:500] 
```
