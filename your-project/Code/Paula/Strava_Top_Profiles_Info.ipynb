```python
#importing libraries

import pandas as pd
import requests
import re 
import json 
from bs4 import BeautifulSoup as Soup
```

```python
#Since we have the problem of data privacy in Strava, let's try to get the names and IDs of the top 100 pros in Strava
#This will give us hopefully enough info to perform our analysis
#We need to grab their name, their id and their data preferences
```

```python
url = "https://www.strava.com/pros"

#let's request the info
page = requests.get(url)
soup = Soup(page.content, "html.parser")
```

```python
top_profiles = [element.get_text() for element in soup.select("div ul.pros")[0].select("h3 a.minimal")]
```

```python
top_profiles_url = [element['href'] for element in soup.select("div ul.pros")[0].select("h3 a.minimal")]
```

```python
top_profiles_url
```

```python
#We now have the names of the pro athletes and their url, let's look for their id and their data rights.
#First, we complete the urls with the remaining part of the https_ : https://www.strava.com

absolute_url = []

for element in top_profiles_url:
    url_root = "https://www.strava.com"
    absolute_url.append(url_root + element)
```

```python
#Now we retrieve the id for each url 

#let's try with the first one
url_1= absolute_url[0]
page_1 = requests.get(url_1)
soup_1 = Soup(page_1.content, "html.parser")

#Let's call the href where the id is located and clean the link so we only weep the number id
id_profile = soup_1.select("div.spans5 a.button")[0]["href"]
id_profile
re.findall(r"(?<=content=)\d*", id_profile) 
#now we find a pattern to retrieve all the id numbers, let's do it for each url in absolute_url
```

```python
trial = absolute_url[0:500]
```

```python
list_ids = []
#We create an empty list to populate with all the ids and retrieve each url to find their number like above
for soup in trial:
    url = soup
    page = requests.get(url)
    soup_1 = Soup(page.content, "html.parser")
    id_profile = soup_1.select("div.spans5 a.button")[0]["href"]
    id_finale = re.findall(r"(?<=content=)\d*", id_profile)
    list_ids.append(id_finale)
```
