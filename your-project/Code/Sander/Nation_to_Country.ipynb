# Athletics

```python
df = pd.read_csv("../../Sources/athletics_clean.csv", sep=",")
df
```

```python
# Only 3 digits for nation
list_alpha_3 = [i.alpha_3 for i in list(pycountry.countries)]
```

```python
def country_calc(df):
    if (len(df['nationality'])==3 and df['nationality'] in list_alpha_3):
        return pycountry.countries.get(alpha_3=df['nationality']).name
    else:
        return 'Invalid Code'
```

```python
df['country'] = df.apply(country_calc, axis= 1)
df
```

```python
# to csv
athletics_country = df.to_csv("../../Sources/atl_country.csv", sep=",")
```
