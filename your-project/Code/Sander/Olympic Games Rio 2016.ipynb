```python
# import libraries
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../../Sources/athletes.csv", sep="|")
df
```

```python
df.info()
```

```python
df = df.drop('info', axis= 1)
```

```python
# Make a list of all competitors of cycling and athletics
new_list =[]
for athlete in range(len(df)):
    if (df['sport'][athlete] == 'cycling')| (df['sport'][athlete] == 'athletics'):
        new_list.append([df['id'][athlete], df['name'][athlete],df['nationality'][athlete],
                         df['sex'][athlete],df['date_of_birth'][athlete],df['height'][athlete],df['weight'][athlete],
                        df['sport'][athlete],df['gold'][athlete],df['silver'][athlete],df['bronze'][athlete]])
        
new_list
```

```python
# make dataframa again
# column names , df.columns
columns = ['id', 'name', 'nationality', 'sex', 'date_of_birth', 'height', 'weight',
       'sport', 'gold', 'silver', 'bronze']

# create df
athletics_and_cycling = pd.DataFrame(new_list, columns= columns)
athletics_and_cycling
```

```python
# data cleaning
athletics_and_cycling.columns
```

```python
# data cleaning
athletics_and_cycling.reset_index(drop=True, inplace=True)
athletics_and_cycling
```

```python
# dataframe wiht only athletics
ath_and_cy = athletics_and_cycling.copy()
athletics = ath_and_cy.loc[ath_and_cy['sport'] == 'athletics']
athletics
```

```python
# missing values for height, weight 166 and 162 respectively
athletics.info()
```

```python
athletics['height'].fillna(athletics['height'].mean(),inplace=True)
```

```python
athletics['weight'].fillna(athletics['weight'].mean(),inplace=True)
```

```python
athletics.info()
```
